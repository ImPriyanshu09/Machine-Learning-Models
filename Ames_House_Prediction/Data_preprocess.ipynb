***Handling Missing Values***

```python
df.head()
#df.info()
```

```python
df = df.drop('PID',axis = 1)
```

```python
df.head()
```

```python
### define a function to get the percentage value for the missing values
def percent_missing(df):
    percent_nan = 100*df.isnull().sum()/len(df) 
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    return percent_nan
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
plt.figure(figsize = (10,6),dpi = 100)
sns.barplot(x = percent_nan.index,y = percent_nan)
plt.xticks(rotation = 70);
```

```python
percent_nan[percent_nan < 1]  ## lets check out the percentage value below 1 %
```

```python
### lets check if the row missing for Electrical is also missing for Garage Area or not
df[df['Electrical'].isnull()]['Garage Area']
```

```python
### the rows are different for both the features where value is missing 
```

```python
df[df['Bsmt Half Bath'].isnull()]
```

```python
df = df.dropna(subset=['Electrical','Garage Cars'],axis=0)
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan[percent_nan<1]
```

```python
plt.figure(figsize = (10,6),dpi = 100)
sns.barplot(x = percent_nan.index,y = percent_nan)
plt.xticks(rotation = 70);
plt.ylim(0,1)
```

```python
### some of the values are not nan but are giving a nan value
## for example the houses having unfinished basement is giving a nan value 
### so now we have to convert that into either 0 or None (for numeric and for string respectively)
```

```python
### BSMT Numeric Columns ------> fillna (0)

bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)
```

```python
#### BSMT String Columns -------> filna (None)

bsmt_str_cols =  ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_str_cols] = df[bsmt_str_cols].fillna('None') 
```

```python
df[df['Bsmt Full Bath'].isnull()]
```
